```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#https://drive.google.com/file/d/1XFViS21mLmAnVUqzVsMYgxZUPw6sYn00/view?usp=sharing
url = "https://drive.google.com/uc?export=download&id=1XFViS21mLmAnVUqzVsMYgxZUPw6sYn00"

inputDF = pd.read_csv(url, sep = ',')

inputDF.head()
```

```python
inputDF.tail()
```

```python
axes = inputDF.plot(marker=None, figsize=(20, 20), subplots = True)
```

```python
matrix = inputDF.values

fig, ax = plt.subplots(1, 1, figsize = (14, 8))

timestamps = range(matrix.shape[0])

begin = 0

step = 50

ax.plot(timestamps[begin : : step], (matrix[begin : : step, 0] + matrix[begin : : step, 1]/1000) - matrix[begin : : step, 2], 'black', marker = '.', linewidth = 2, markersize = 5)

ax.legend(['Разница выработки и потребления, МВ'])

plt.grid(True)
plt.show()

energy = 0
for i in timestamps:
  energy += (matrix[i, 0] + matrix[i, 1]/1000) - matrix[i, 2]
print(f'Количество избыточной энергии: {energy} МВт')
```
